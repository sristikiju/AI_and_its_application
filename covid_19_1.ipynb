# **Outlier Detection and Handling**

```python
# Drop 'Id' column if it exists
df_copy = df.drop('Id', axis=1).copy() if 'Id' in df.columns else df.copy()

# Clean column names
df_copy.columns = df_copy.columns.str.strip()

# Detect and remove outliers using IQR for each numeric column
for column in df_copy.select_dtypes(include='number').columns:
    q1 = df_copy[column].quantile(0.25)
    q3 = df_copy[column].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    df_copy = df_copy[(df_copy[column] >= lower_limit) & (df_copy[column] <= upper_limit)]

# Select a few numeric columns relevant for boxplot
selected_columns = ["AGE"]
df_plot = df_copy[selected_columns]

# Plot boxplot after removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_plot, palette="Set3")

# Set labels and title
plt.title("Boxplot of COVID-19 Dataset Features (Outliers Removed)", fontsize=16)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Values", fontsize=12)
# plt.xticks(rotation=45)

# Show plot
# plt.tight_layout()
plt.show()
```
